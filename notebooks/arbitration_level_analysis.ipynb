# Arbitration-Level Analysis

Visualizations for the arbitration-level dataset produced by `build_datasets.py`. Focuses on redaction counts, page counts, time series trends by year, and classification breakdowns.

```python
# Cell 2: Imports & Configuration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
PALETTE = "Set2"
DEFAULT_FIGSIZE = (10, 6)
WIDE_FIGSIZE = (14, 5)
HIST_KWARGS = dict(bins=30, edgecolor="black", alpha=0.7)
```

```python
# Cell 3: Load Data
# >>> UPDATE THIS PATH to point to your arbitration-level CSV <<<
DATA_PATH = "../datasets/arbitration_level.csv"

df = pd.read_csv(DATA_PATH)
print(f"Shape: {df.shape}")
print()
df.dtypes
```

## Time Series (by year)

```python
# Cell 4: Total detections over time
yearly_detections = df.groupby("year")["total_detections"].sum()

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
yearly_detections.plot.bar(ax=ax, color=sns.color_palette(PALETTE)[0], edgecolor="black")
ax.set_title("Total Detections Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Sum of Total Detections")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cell 5: Total pages over time
yearly_pages = df.groupby("year")["total_pages"].sum()

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
yearly_pages.plot.bar(ax=ax, color=sns.color_palette(PALETTE)[1], edgecolor="black")
ax.set_title("Total Pages Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Sum of Total Pages")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Cell 6: Mean fraction of pages with redactions over time
yearly_fraction = df.groupby("year")["fraction_pages_with_redactions"].mean()

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
yearly_fraction.plot(ax=ax, marker="o", color=sns.color_palette(PALETTE)[2])
ax.set_title("Mean Fraction of Pages with Redactions Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Mean Fraction")
ax.set_ylim(bottom=0)
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Mean coverage percent over time
yearly_coverage = df.groupby("year")["mean_coverage_percent"].mean()

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
yearly_coverage.plot(ax=ax, marker="o", color=sns.color_palette(PALETTE)[3])
ax.set_title("Mean Coverage Percent Over Time")
ax.set_xlabel("Year")
ax.set_ylabel("Mean Coverage (%)")
ax.set_ylim(bottom=0)
plt.tight_layout()
plt.show()
```

## Distributions

Histograms are filtered to rows where the plotted value > 0 to avoid a large spike at zero.

```python
# Cell 8: Distribution of total_detections per arbitration
data = df.loc[df["total_detections"] > 0, "total_detections"]

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
ax.hist(data, **HIST_KWARGS, color=sns.color_palette(PALETTE)[0])
ax.set_title("Distribution of Total Detections per Arbitration (>0)")
ax.set_xlabel("Total Detections")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Cell 9: Distribution of total_pages per arbitration
data = df.loc[df["total_pages"] > 0, "total_pages"]

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
ax.hist(data, **HIST_KWARGS, color=sns.color_palette(PALETTE)[1])
ax.set_title("Distribution of Total Pages per Arbitration (>0)")
ax.set_xlabel("Total Pages")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Cell 10: Distribution of fraction_documents_with_redactions
data = df.loc[df["fraction_documents_with_redactions"] > 0, "fraction_documents_with_redactions"]

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
ax.hist(data, **HIST_KWARGS, color=sns.color_palette(PALETTE)[2])
ax.set_title("Distribution of Fraction of Documents with Redactions (>0)")
ax.set_xlabel("Fraction")
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Cell 11: Scatter — total_pages vs total_detections (colored by decade)
df["decade"] = (df["year"] // 10) * 10

fig, ax = plt.subplots(figsize=DEFAULT_FIGSIZE)
for decade, group in df.groupby("decade"):
    ax.scatter(group["total_pages"], group["total_detections"],
               label=f"{decade}s", alpha=0.6, edgecolors="black", linewidths=0.5)
ax.set_title("Total Pages vs Total Detections (colored by decade)")
ax.set_xlabel("Total Pages")
ax.set_ylabel("Total Detections")
ax.legend(title="Decade")
plt.tight_layout()
plt.show()
```

## Classification Breakdown by Year

```python
# Cell 12: Stacked bar — classification breakdown by year
class_cols = ["total_inline", "total_partial_line", "total_full_line", "total_block"]
yearly_class = df.groupby("year")[class_cols].sum()

fig, ax = plt.subplots(figsize=WIDE_FIGSIZE)
yearly_class.plot.bar(stacked=True, ax=ax, colormap="Set2", edgecolor="black")
ax.set_title("Classification Breakdown by Year")
ax.set_xlabel("Year")
ax.set_ylabel("Total Count")
ax.legend(title="Classification", bbox_to_anchor=(1.02, 1), loc="upper left")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
